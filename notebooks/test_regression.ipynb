```python
from pyAudioAnalysis import audioTrainTest as aT
import os
import re
import json
```

```python
folder_tag = 'midi_audio_sample20180819013100'
model_types = ['svm', 'svm_rbf', 'randomforest']
data_folder = './{0}/testing'.format(folder_tag)
regression_result_json = './{0}/testing/regression_result.json'.format(folder_tag)
```

```python
files = [f for f in os.listdir(data_folder) if re.match('.*\.ogg$', f)]
#files
```

```python
test_results = []
```

```python
for file_model in [(data_folder + '/' + f, f, './{0}/{1}'.format(folder_tag, model_type), model_type) for f in files for model_type in model_types]:
    file_name_full_path = file_model[0]
    file_name = file_model[1]
    model_suffix = file_model[2]
    model_type = file_model[3]
    print('file:', file_name, 'model_type:', model_type)
    
    m = re.match('^note-program(?P<program>\d+)-degree(?P<degree>\d+)-volume(?P<volume>\d+).*.ogg$', file_name)
    v_program = m.group('program')
    v_degree = m.group('degree')
    v_volume = m.group('volume')
    
    regression = aT.fileRegression(file_name_full_path, model_suffix, model_type)
    
    #print(test_result)
    top = [(regression[0][i], regression[1][i]) for i in range(0, len(regression[0]), 1) if regression[0][i] == max(regression[0])][0]
    expected_program = 'program{0}'.format(v_program)
    is_success = top[1] == expected_program
    #print('success: ', is_success, 'expected:', expected_program)
    
    test_results.append({ 'success': is_success, 'model type': model_type, 'audio file': file_name, 'degree':v_degree, 'volume':v_volume, 'regression': regression })
```

```python
print(test_results)
with open(regression_result_json, 'w') as outfile:
    json.dump(test_results, outfile)
```
